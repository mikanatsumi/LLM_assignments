```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "microsoft/Phi-3.5-mini-instruct"
api_token = "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"
# Load tokenizer and model with authentication
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=api_token)
model =AutoModelForCausalLM.from_pretrained(model_name, use_auth_token=api_token)

import torch
from datasets import load_dataset
import re

def extract_answer_after_prompt(generated_text):
    # Define the regular expression pattern to match the text after "Answer:"
    pattern = r"Answer is:\s*(.*)"
    
    # Search for the pattern in the generated text
    match = re.search(pattern, generated_text)
    
    if match:
        # Return the string found after "Answer:"
        match= match.group(1).strip() 
        cleaned_answer = re.sub(r"^\d[\.\)]\s*",  "",match)  # Remove '1)', '2)', '3)', '4)'
        
        return cleaned_answer.strip()  # Return the cleaned answer, with leading/trailing whitespace removed# Strip leading/trailing whitespace
    else:
        # If no match is found, return None or handle it appropriately
        return None

# Load the mathematics question-answer dataset
dataset = load_dataset("cais/mmlu", "college_mathematics")
device='cuda:0'
model=model.to(device)
# A function to format the question and options
def create_react_prompt(question, options):
    prompt = f"Q: {question}\n"
    prompt += "Please follow these steps:\n"
    prompt += "1. Reason step-by-step to understand the problem.\n"
    prompt += "2. Suggest an action (e.g., performing a calculation) if needed.\n"
    prompt += "\nOptions:\n"
    for i, option in enumerate(options, 1):
        prompt += f"{i}. {option}\n"
    prompt += "Answer is:"
    return prompt
# Example use


# Perform zero-shot classification for each question in the dataset
correct_predictions = 0
total_questions = 0

for sample in dataset["test"]:
    question = sample["question"]
    options = sample["choices"]
    correct_answer = sample["answer"]

    # Prepare the input for the model
    input_text = create_react_prompt(question, options)
    inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True).to(device)

    # Generate model outputs
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=20)
    
    # Decode the generated text
    predicted_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
   
    predicted_answer=extract_answer_after_prompt(predicted_text)
    
    correct_answer = sample["choices"][sample["answer"]]
    print(predicted_answer,correct_answer)
    if predicted_answer == correct_answer:
        correct_predictions += 1
        print('correct')
    total_questions += 1

    # Print progress
    if total_questions % 100 == 0:
        print(f"Processed {total_questions} questions")

# Calculate accuracy
accuracy = correct_predictions / total_questions
print(f"Accuracy: {accuracy * 100:.2f}%")
```
